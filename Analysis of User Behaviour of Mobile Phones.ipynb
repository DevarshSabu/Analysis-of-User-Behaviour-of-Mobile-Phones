### Visualization

#### Boxplot

```python
n_dm=df['Device Model'].value_counts()
n_dm=pd.DataFrame(n_dm)
n_dm
```

```python
dm=df['Device Model'].unique()
pd.DataFrame(dm)
```

```python
plt.figure(figsize=(10, 3))
plt.bar(dm, n_dm['count'],color=['red', 'orange', 'yellow', 'green', 'grey'])
plt.title("Bar Plot showing the Models") 
plt.xlabel("Device model")
plt.ylabel("Number of Users")
plt.yticks(np.arange(0,160,10))
plt.xticks( rotation=25)
plt.show()
```

#### Scatter plot

```python
plt.figure(figsize=(10, 3))
plt.scatter(df['App Usage Time (min/day)'],df['Screen On Time (hours/day)'],color='red',sizes=(1, 50))
plt.title('Scatter plot showing the app usage time and screen on time')
plt.xlabel('App Usage Time (min/day)')
plt.ylabel('Screen On Time (hours/day)')
plt.xticks(np.arange(0, 650, 50))
plt.show(close=None, block=None)
```

```python
plt.figure(figsize=(10, 3))
sns.pairplot(df)
plt.show(close=None, block=None)
```

#### Histogram

```python
continous_features = [feature for feature in df.columns if df[feature].dtypes!="O" and df[feature].nunique()>90]
continous_features

categorical_features = [feature for feature in df.columns if df[feature].dtypes=='O']
categorical_features

for i in continous_features:
    plt.figure(figsize=(10, 3))
    plt.hist(df[i], )
    plt.title(i)
    plt.xlabel(i)
    plt.ylabel('Number of users')
    plt.show()
```

#### Box Plot

```python
continous_features = [feature for feature in df.columns if df[feature].dtypes!="O" and df[feature].nunique()>90]
continous_features

categorical_features = [feature for feature in df.columns if df[feature].dtypes=='O']
categorical_features

for i in continous_features:
    plt.figure(figsize=(10, 3))
    plt.boxplot(df[i],vert=False)
    plt.title(i)
    plt.show()
```

#### Heatmap

```python
corr_matrix = df.corr(numeric_only = True)
corr_matrix 
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(corr_matrix,vmin = -1,vmax = 1,annot =True)
```
